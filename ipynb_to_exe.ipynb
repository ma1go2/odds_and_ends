# 주피터노트북(.ipynb) 파일을 실행파일로 만들기

우선 pyinstaller를 설치한다.

```python
pip install pyinstaller
```

실행파일로 만들고 싶은 파일을 py로 저장한다.
> md_rename_today.py는 Markdown파일 이름 앞에 오늘의 날짜를 추가한다.



파일 위치에서 터미널을 열어 실행

`pyinstaller -w -F --icon=icon.ico .\md_rename_today.py`  
아이콘은 [icoconvert](https://icoconvert.com/)사이트를 이용하여 변환하면 된다.  
아이콘과 py파일은 같은 위치에 두거나 경로를 지정해서 설정하면 된다.  

파일의 SPEC파일, dist, build 폴더가 생긴다.  
dist폴더안에 실행 파일(exe)가 생성된다.
